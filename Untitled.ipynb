### Preprocess data

```python
# As we see, the useful columns are all have invalid data type (Object). Now we convert it to number / category type
useful_cols = list(set(useful_cols) - set(['Rated', 'Production']))
# Let see the categories of each columns:
for useful_col in useful_cols:
    print(useful_col)
    print(data_df[useful_col].unique())
    print()
```

```python
# First of all, we can see that there is a column whose type could be category but we have to preprocess it a little.
# That column is: Genre

data_df['Genre'].unique()

all_gernes_in_one_list = []
for gerne in data_df['Genre']:
    if not isinstance(gerne, float):
        all_gernes_in_one_list.extend(gerne.split(', '))
    
all_gernes_in_one_list_unique = list(set(all_gernes_in_one_list))

gerne_percentages_dict = {}
for gerne in all_gernes_in_one_list_unique:
    gerne_percentages_dict[gerne] = round(all_gernes_in_one_list.count(gerne)/len(all_gernes_in_one_list) * 100, 1)

dict(sorted(gerne_percentages_dict.items(), key=lambda item: item[1]))

# OK, we can see that the columns with category data type dont diverse significantly. We can use this column for
# prediction.
```
